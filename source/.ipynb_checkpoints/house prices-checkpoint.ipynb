### Jak kształtuje się cena mieszkania dla Krakowa


```python
import pandas 
import numpy as np

cracov_df = df.loc[df['city_name'].str.lower() == "kraków"]

bins = np.arange(start=0, step=25000, stop=3000000)
ax = cracov_df.plot.hist(column='price', grid=True, color='#607c8e', bins = bins)
ax.set_xlabel("ceny mieszkań [tyś]")
ax.set_ylabel("ilość mieszkań")

cracov_prices = cracov_df['price']

data = {
    'mean': [cracov_prices.mean()],
    '75 %': [cracov_prices.quantile(0.75)],
    '90 %': [cracov_prices.quantile(0.90)],
    '95 %': [cracov_prices.quantile(0.95)],
    '99 %': [cracov_prices.quantile(0.99)],
    'std': [cracov_prices.std()],
}


df_cracov_prices_stats = pandas.DataFrame(data = data)

display(df_cracov_prices_stats)
```

### Test Shapiro-Wilk dla cen mieszkań w Krakowie

Czy ceny mają rozkład normalny?

HO - Próba pochodzi z populacji o rozkładzie normalnym

H1 - Próba nie pochodzi z populacji o rozkładzie normalnym.

*n* = 50

*threashold* = 5%

[shapiro doc](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html)

```python
from scipy.stats import shapiro

sample = cracov_df[['price']].sample(50)

stest = shapiro(sample)
display(stest)

if (stest.pvalue < 0.05):
    print("pvalue {} < {}".format(stest.pvalue, 0.05))
    print("Hipoteza H0 odrzucona - próbka nie pochodzi z rozkładu normalnego")
else:
    print("pvalue {} > {}".format(stest.pvalue, 0.05))
    print("Hipoteza H0 nie może być odrzucona - próbka może pochodzić z rozkładu normalnego")
```

### Test Kołmogorowa dla cen mieszkań w Krakowie

Czy ceny mają rozkład normalny?

HO - Próba pochodzi z populacji o rozkładzie normalnym

H1 - Próba nie pochodzi z populacji o rozkładzie normalnym.

*n* = 1000

*threashold* = 5%

[ktest doc](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kstest.html)

```python
from scipy.stats import kstest
import numpy as np

sample = cracov_df[['price']].sample(1000)
flat_prices = [value[0] for value in sample[['price']].values]
np.array(flat_prices)
normalized = flat_prices/np.linalg.norm(flat_prices)

ktest = kstest(normalized, 'norm')
display(ktest)

if (stest.pvalue < 0.05):
    print("pvalue {} < {}".format(stest.pvalue, 0.05))
    print("Hipoteza H0 odrzucona - próbka nie pochodzi z rozkładu normalnego")
else:
    print("pvalue {} > {}".format(stest.pvalue, 0.05))
    print("Hipoteza H0 nie może być odrzucona - próbka może pochodzić z rozkładu normalnego")
```

### Czy isnieje zależność mięczy wielkością miasta a ceną mieszkania za m2
 

```python
# define bins for population and price per meter
population_bin = [0, 1000, 2500, 5000, 10000, 25000, 50000, 100000, 250000, 500000, 1000000, 5000000]
price_per_meter_bin = [0, 3000, 5000, 7500, 10000, 12500, 15000, 17500, 20000, 22500, 25000, 30000, 50000]

# define new data frame
df_population_price = pandas.DataFrame()
df_population_price['City Population'] = pandas.cut(df['population'], population_bin)
df_population_price['Price / m2'] = pandas.cut(df['price_per_meter'], price_per_meter_bin)

# create contingency table for city population and price per sq meter  
data_crosstab = pandas.crosstab(df_population_price['City Population'],
                            df_population_price['Price / m2'],
                           margins=True, margins_name="Total")

```

Poniżej znajduje się tabela, która ilość ofert mieszkań dla populacji miast, oraz cenę za metr kwadratow

```python
display(data_crosstab)
```

#### Test Chi kwadrat - czy isnieje zależność między wielkością miasta a ceną mieszkania za m2


```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2

stat, p, dof, expected = chi2_contingency(data_crosstab)

# test-statistic
prob = 0.95
critical = chi2.ppf(prob, dof)
print('probability=%.2f, critical=%.2f, stat=%.2f' % (prob, critical, stat))

if abs(stat) >= critical:
    display('Hipoteza H0 odrzucona - istnieje zależność między pouplacją miasta, a ceną za m2')
else:
    display('Hipoteza H0 nie może być odrzucona - nie istnieje zależność między pouplacją miasta, a ceną za m2')

# p-value
alpha = 1.0 - prob
display('significance=%.2f, p=%.2f' % (alpha, p))

if p <= alpha:
    display('Hipoteza H0 odrzucona - istnieje zależność między pouplacją miasta, a ceną za m2')
else:
    display('Hipoteza H0 nie może być odrzucona - nie istnieje zależność między pouplacją miasta, a ceną za m2')
```

### Regresja liniowa między wielkością miasta a ceną mieszkania za m2


```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

population_bin = [1000, 2500, 5000, 7500, 10000, 12500, 15000, 17500, 20000, 25000, 50000, 100000, 250000, 500000, 750000, 2000000]
price_per_meter_bin = [0, 3000, 5000, 7500, 10000, 12500, 15000, 17500, 20000, 22500, 25000, 30000, 50000]

df_population_price = pandas.DataFrame()
df_population_price['City Population'] = pandas.cut(df['population'], population_bin)
df_population_price['Price / m2'] = df['price_per_meter']
df_avg_price_by_population = df_population_price[['City Population', 'Price / m2']].groupby('City Population').mean().reset_index()
display(df_avg_price_by_population)

df_avg_price_by_population['City Population'] = df_avg_price_by_population['City Population'].map(lambda x: x.right)

plt.scatter(df_avg_price_by_population['City Population'], df_avg_price_by_population['Price / m2'])
plt.xlabel('górny przedział populacji miasta w tyś')
plt.ylabel('średnia cena za m2')
plt.title('Średnia cena mieszkania na m2 dla przedziałach populacj')

population = df_avg_price_by_population['City Population'].values.reshape(-1,1)
price = df_avg_price_by_population['Price / m2'].values.reshape(-1,1)
lr = LinearRegression()
lr.fit(population, price)
price_pred = lr.predict(population)
plt.plot(population, price_pred, color='red')

plt.show()
```
